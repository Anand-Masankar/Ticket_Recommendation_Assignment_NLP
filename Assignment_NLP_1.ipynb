-- we get a sparse matrix as "word2vec_model" now converting it into dense matrix

````python
word2vec_model.vector_size
````

````python
# now we have to create dense matrix for word2vec model, because whatever we get above is sparse matrix

def w2v_vector(complaint_list, w2v_model):
    complaint_vector = []
    for complaint in complaint_list:
        zero_vec = np.zeros(w2v_model.vector_size)
        matrix = []
        
        for word in complaint:
            if word in w2v_model.wv:
                try :
                    matrix.append(w2v_model.wv[word])
                except KeyError:
                    continue
        if matrix:
            matrix = np.asarray(matrix)
            mean_vector = matrix.mean(axis = 0)
            
            complaint_vector.append(mean_vector)
        else:
            complaint_vector.append(zero_vec)
            
    return complaint_vector
 
````
